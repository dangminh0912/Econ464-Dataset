```python
# Import libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
# Data
file_list = ["/Users/dangminh0912/Desktop/Year 4 (Sem 2)/ECON 464/Final Project/Data set/nba_salary.html",
             "/Users/dangminh0912/Desktop/Year 4 (Sem 2)/ECON 464/Final Project/Data set/wnba_salary.html"]
```

```python
# League = 0: NBA; League = 1: WNBA
def web_scrap(league):
    
    # Variables
    rank_list = []
    name_list = []
    position_list = []
    age_list = []
    average_list = []
    
    # Web-scrapping
    with open(file_list[league], 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Parse the HTML
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the table
    table = soup.find('table', {'class': 'datatable noborder'})

    # Find the rows in the table
    rows = table.find_all('tr')
        
    for row in rows: 
        cells = row.find_all('td')
        if len(cells) >= 2:
            rank = cells[0].text.strip()
            name = cells[1].find('h3').text.strip()  # Extract the name from the <h3> tag
            position = cells[2].text.strip()
            age = cells[3].text.strip()
            average = cells[4].text.strip()
                    
            rank_list.append(rank)
            name_list.append(name)
            position_list.append(position)
            age_list.append(age)
            average_list.append(average)
            
    
    # Create dataframe
    df = pd.DataFrame({
        'Rank': rank_list,
        'Player': name_list,
        'Position': position_list,
        'Age': age_list,
        'Average': average_list
    })
    
    return df

  
      
```

```python
df1 = web_scrap(0)
excel_path = 'nba_wage.xlsx'
df1.to_excel(excel_path, index=False)
df1
```

```python
df2 = web_scrap(1)
excel_path = 'wnba_wage.xlsx'
df2.to_excel(excel_path, index=False)
df2
```

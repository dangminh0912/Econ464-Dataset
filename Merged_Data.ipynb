```python
import pandas as pd
```

```python
# Read the Excel file into a pandas DataFrame
nba_salary = pd.read_excel('nba_wage (1).xlsx') 
wnba_salary = pd.read_excel('wnba_wage (1).xlsx')
```

```python
nba_salary = nba_salary[['NAME','SALARY']]
nba_salary
```

```python
wnba_salary= wnba_salary[['Name','Salary']]
wnba_salary
```

```python
nba_salary.rename(columns={'NAME': 'Name', 'SALARY': 'Salary'}, inplace=True)
```

```python
# Add a 'Gender' column to each DataFrame
nba_salary['Gender'] = 1  # NBA gender code
wnba_salary['Gender'] = 0  # WNBA gender code

# Concatenate the DataFrames vertically
salary = pd.concat([nba_salary, wnba_salary], ignore_index=True)
```

```python
salary 
```

```python
# Read the Excel file into a pandas DataFrame
nba_stats = pd.read_excel('NBA-All-Player (1).xlsx') 
wnba_stats = pd.read_excel('wnba_stats.xlsx')
```

```python
nba_stats = nba_stats[['PLAYER','TEAM','AGE','GP','W','L','MIN','PTS','+/-']]
nba_stats 
```

```python
wnba_stats = wnba_stats[['PLAYER','TEAM','AGE','GP','W','L','MIN','PTS','+/-']]
wnba_stats 
```

```python
stat = pd.concat([wnba_stats, nba_stats], ignore_index=True)
```

```python
stat
```

```python
salary.rename(columns={'Name': 'PLAYER'}, inplace=True)
```

```python
data = pd.merge(stat, salary, on='PLAYER', how='inner')
```

```python
data
```

```python
gender_counts = data['Gender'].value_counts()
```

```python
gender_counts.to_csv('merged_data.csv')
```
